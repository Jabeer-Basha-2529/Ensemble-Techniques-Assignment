Question 1: What is Ensemble Learning in Machine Learning? Explain the key idea behind it.
- Ensemble Learning is a machine learning technique in which multiple models are combined to produce a better prediction than a single model.

- The main idea is that different models make different mistakes. By combining their predictions, the overall model becomes more accurate, stable, and robust.

- Benefits:
   - Improves prediction accuracy
   - Reduces overfitting
   - Handles noisy data better
   - Produces more reliable results

- Examples:
  - Random Forest
  - Bagging
  - Boosting
  - AdaBoost
  - Gradient Boosting
  - XGBoost

Question 2: What is the difference between Bagging and Boosting?

|Bagging                     | Boosting                                      |
| --------------------------- | --------------------------------------------- |
| Builds models independently | Builds models sequentially                    |
| Uses bootstrap sampling     | Focuses on correcting previous errors         |
| Reduces variance            | Reduces bias                                  |
| Models run in parallel      | Models depend on previous models              |
| Less prone to overfitting   | Can overfit if not tuned                      |
| Example: Random Forest      | Example: AdaBoost, Gradient Boosting, XGBoost |

Question 3: What is bootstrap sampling and what role does it play in Bagging methods like Random Forest?
- Bootstrap sampling is a technique where random samples are drawn from the original dataset with replacement.

 Because sampling is with replacement:

  - Some records appear multiple times.
  - Some records are not selected.

 Each Decision Tree in Random Forest is trained using a different bootstrap sample.

 This helps:

  - Increase diversity among trees.
  - Reduce variance.
  - Improve generalization.
  - Reduce overfitting.


Question 4: What are Out-of-Bag (OOB) samples and how is OOB score used to evaluate ensemble models?
- Out-of-Bag (OOB) samples are the data points that are not selected during bootstrap sampling.

 Approximately 36.8% of the training data remains unused for each tree.

 These samples are used to test that tree.

 The average prediction accuracy over all OOB samples is called the OOB Score.

 Advantages:

   - No separate validation dataset required.
   - Saves training data.
   - Gives a reliable estimate  of model performance.

| Decision Tree                               | Random Forest                               |

 Random Forest generally provides more reliable feature importance because it averages the importance across multiple trees.



```python
# Question 6: Python Program – Random Forest Feature Importance
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
import pandas as pd

# Load dataset
data = load_breast_cancer()

X = data.data
y = data.target

# Train model
model = RandomForestClassifier(random_state=42)
model.fit(X, y)

# Feature importance
importance = pd.DataFrame({
    "Feature": data.feature_names,
    "Importance": model.feature_importances_
})

importance = importance.sort_values(
    by="Importance",
    ascending=False
)

print("Top 5 Important Features:\n")
print(importance.head())
```

```python
# Question 7: Bagging Classifier vs Decision Tree
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import accuracy_score

# Load data
iris = load_iris()

X_train, X_test, y_train, y_test = train_test_split(
    iris.data,
    iris.target,
    test_size=0.2,
    random_state=42
)

# Decision Tree
tree = DecisionTreeClassifier(random_state=42)
tree.fit(X_train, y_train)

pred1 = tree.predict(X_test)

# Bagging
bag = BaggingClassifier(
    estimator=DecisionTreeClassifier(),
    n_estimators=50,
    random_state=42
)

bag.fit(X_train, y_train)

pred2 = bag.predict(X_test)

print("Decision Tree Accuracy:",
      accuracy_score(y_test, pred1))

print("Bagging Accuracy:",
      accuracy_score(y_test, pred2))
```

```python
# Question 8: Random Forest with GridSearchCV
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# Load dataset
data = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(
    data.data,
    data.target,
    test_size=0.2,
    random_state=42
)

param_grid = {
    "n_estimators":[50,100,200],
    "max_depth":[3,5,10,None]
}

grid = GridSearchCV(
    RandomForestClassifier(random_state=42),
    param_grid,
    cv=5
)

grid.fit(X_train, y_train)

print("Best Parameters:", grid.best_params_)

print("Test Accuracy:",
      grid.score(X_test,y_test))
```

```python
# Question 9: Bagging Regressor vs Random Forest Regressor
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

# Load dataset
housing = fetch_california_housing()

X_train, X_test, y_train, y_test = train_test_split(
    housing.data,
    housing.target,
    test_size=0.2,
    random_state=42
)

# Bagging
bag = BaggingRegressor(
    estimator=DecisionTreeRegressor(),
    random_state=42
)

bag.fit(X_train, y_train)

pred1 = bag.predict(X_test)

# Random Forest
rf = RandomForestRegressor(random_state=42)

rf.fit(X_train, y_train)

pred2 = rf.predict(X_test)

print("Bagging MSE:",
      mean_squared_error(y_test,pred1))

print("Random Forest MSE:",
      mean_squared_error(y_test,pred2))
```

Question 10: Loan Default Prediction Using Ensemble Learning
- To predict loan defaults, I would follow these steps:

  1. Understand the Problem
     - Define the target variable (loan default: Yes/No).
     - Identify features such as customer age, income, credit score, transaction history, and loan amount.
  2. Choose Between Bagging and Boosting
    - Use Bagging (Random Forest) if the goal is to reduce overfitting and improve stability.
    - Use Boosting (e.g., Gradient Boosting or XGBoost) if achieving the highest prediction accuracy is the priority by correcting previous errors.
  3. Handle Overfitting
     - Limit tree depth.
     - Use cross-validation.
     - Tune hyperparameters such as the number of trees, learning rate (for boosting), and maximum depth.
  4. Select Base Models
     - Decision Trees are commonly used as base learners because they capture nonlinear relationships and work well in ensemble methods.
  5. Evaluate Performance
     - Split the data into training and testing sets.
     - Use k-fold cross-validation.
     - Measure performance using Accuracy, Precision, Recall, F1-score, and ROC-AUC.
  6. Why Ensemble Learning Improves Decision-Making
     - Combines predictions from multiple models, reducing errors from any single model.
     - Produces more stable and accurate predictions.
     - Better identifies high-risk loan applicants, helping financial institutions reduce defaults and make informed lending decisions.
